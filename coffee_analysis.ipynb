```python
# Import essential libraries
import numpy as np          # For numerical operations
import pandas as pd         # For data manipulation
import matplotlib.pyplot as plt  # For plotting
import seaborn as sns       # For statistical visualizations
```

```python
# Load the Excel dataset
file_path = 'coffee_analysis.xlsx'  # Make sure the file is in the same folder as your notebook
df = pd.read_excel(file_path) 
```

```python
# 1. First 5 rows
print("1. First 5 Rows of the Dataset:")
display(df.head())

# 2. Last 5 rows
print("\n2. Last 5 Rows of the Dataset:")
display(df.tail())

# 3. Shape of the dataset
print(f"\n3. Dataset Shape: {df.shape[0]} rows, {df.shape[1]} columns")

# 4. Column names
print("\n4. Column Names:")
print(df.columns.tolist())

# 5. Data types
print("\n5. Data Types:")
print(df.dtypes)
```

```python
# 6. Missing values
print("\n6. Missing Values in Each Column:")
print(df.isnull().sum())

# 7. Summary statistics (numerical)
print("\n7. Summary Statistics (Numerical Columns):")
display(df.describe())

# 8. Summary statistics (categorical)
print("\n8. Summary Statistics (Categorical Columns):")
display(df.describe(include=[object]))

# 9. Unique values per column
print("\n9. Number of Unique Values per Column:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")

# 10. Number of duplicated rows
print(f"\n10. Number of Duplicated Rows: {df.duplicated().sum()}")

# 11. Count of each data type
print("\n11. Count of Each Data Type:")
print(df.dtypes.value_counts())

```

```python
# Display how many rows will be removed
missing_rows = df.isnull().any(axis=1).sum()
print(f"Removing {missing_rows} rows containing missing values...")

# Remove rows with any missing values
df_cleaned = df.dropna()

# Confirm the shape after removal
print(f"New shape of the dataset: {df_cleaned.shape[0]} rows, {df_cleaned.shape[1]} columns")
```

```python
# Chart 1: Distribution of Coffee Ratings (Histogram + KDE)
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'], kde=True, bins=30, color='saddlebrown')
plt.title('Distribution of Coffee Ratings', fontsize=16)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Chart 2: Average Coffee Rating by Country (Top 10)
top_countries = df.groupby('loc_country')['rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='YlOrBr')
plt.title('Top 10 Countries by Average Coffee Rating', fontsize=16)
plt.xlabel('Average Rating')
plt.ylabel('Country')
plt.xlim(0, 10)  # assuming rating scale is 0–10
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Chart 3: Count of Coffee Roast Types
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='roast', order=df['roast'].value_counts().index, palette='Set2')
plt.title('Distribution of Coffee Roast Types', fontsize=16)
plt.xlabel('Roast Type')
plt.ylabel('Number of Coffees')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Chart 4: Average Price (per 100g in USD) by Roast Type
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='roast', y='100g_USD', estimator='mean', ci=None, palette='Set3')
plt.title('Average Price by Roast Type (per 100g)', fontsize=16)
plt.xlabel('Roast Type')
plt.ylabel('Average Price in USD')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Chart 5 (Alternative): Line Plot of Coffee Reviews Over Time
df['review_date'] = pd.to_datetime(df['review_date'])  # Ensure datetime format
df['year'] = df['review_date'].dt.year

# Group by year and count reviews
yearly_reviews = df.groupby('year').size()

plt.figure(figsize=(10, 6))
sns.lineplot(x=yearly_reviews.index, y=yearly_reviews.values, marker='o', linewidth=2.5, color='darkgreen')
plt.title('Trend of Coffee Reviews Over Years', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(yearly_reviews.index, rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Chart 6: Pie Chart of Coffee Roast Type Distribution
roast_counts = df['roast'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(
    roast_counts.values,
    labels=roast_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette('pastel')
)
plt.title('Distribution of Coffee Roast Types', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures a perfect circle
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='roast', y='rating', palette='coolwarm')
plt.title('Rating Distribution by Roast Type', fontsize=16)
plt.xlabel('Roast Type')
plt.ylabel('Rating')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Create a pivot table for the heatmap
heatmap_data = df.pivot_table(
    index='roast',
    columns='loc_country',
    values='rating',
    aggfunc='mean'
)

# Plot the heatmap with brown shades
plt.figure(figsize=(12, 6))
sns.heatmap(
    heatmap_data,
    annot=True,
    fmt=".1f",
    cmap='copper',    # brown-like colormap
    linewidths=0.5,
    linecolor='white'
)

plt.title('Average Rating by Roast Type and Country', fontsize=16)
plt.xlabel('Country')
plt.ylabel('Roast Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
